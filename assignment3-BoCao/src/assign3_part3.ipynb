```python
'''
Author: Bryan Bo Cao
Email: boca7588@colorado.edu or bo.cao-1@colorado.edu
Github Repo: https://github.com/BryanBo-Cao/neuralnets-deeplearning
'''
from __future__ import print_function

import tensorflow as tf
import random
import copy
import matplotlib.pyplot as plt
import csv
```

```python
test_accuracies = []

# shuffle the data 20 times
for _i in range(20):
    print ("shuffle i: ", _i)
    random.shuffle(dataset)
    train_data = dataset[0:89]
    test_data = dataset[89:104]
    
    # get train input and output data
    train_data_X = []
    train_data_Y = []
    train_data_labels = []
    for i in range(len(train_data)):
        train_data_X.append(train_data[i][1])
        train_data_Y.append(train_data[i][2])
        for j in range(len(train_data[i][2])):
            if (train_data[i][2][j] == 1):
                train_data_labels.append(j)
    #print ("train_data_X: ", train_data_X)
    #print ("train_data_Y: ", train_data_Y)
    print ("train_data_labels: ", train_data_labels)
             
    # get train input and output data
    test_data_X = []
    test_data_Y = []
    test_data_labels = []
    for i in range(len(test_data)):
        test_data_X.append(test_data[i][1])
        test_data_Y.append(test_data[i][2])
        for j in range(len(test_data[i][2])):
            if (test_data[i][2][j] == 1):
                test_data_labels.append(j)
    #print ("test_data_X: ", test_data_X)
    #print ("test_data_Y: ", test_data_Y)
    print ("test_data_labels: ", test_data_labels)
                
    # build the network architecture
    lr = 0.0025
    epochs = 100000
    n_input = 36
    n_H = 12
    n_output = 24

    X = tf.placeholder(tf.float32)
    Y = tf.placeholder(tf.int32)

    wsX_H = tf.Variable(tf.random_uniform([n_input, n_H], -3.0, 3.0))
    wsH_Y = tf.Variable(tf.random_uniform([n_H, n_output], -3.0, 3.0))

    H_outputs = tf.nn.relu(tf.matmul(X, wsX_H))
    hy = tf.nn.relu(tf.matmul(H_outputs, wsH_Y))
    softmax_hy = tf.nn.softmax(hy)

    cost = tf.nn.sparse_softmax_cross_entropy_with_logits(labels = train_data_labels, logits = hy)
    optimizer = tf.train.GradientDescentOptimizer(lr).minimize(cost)

    init = tf.global_variables_initializer()
    
    highest_test_accuracy = 0
    
    with tf.Session() as sess:
        sess.run(init)
        for step in xrange(epochs):
            sess.run(optimizer, feed_dict={X: train_data_X, 
                                           Y: train_data_Y})

            if step % 10000 == 0:
                print (" ")
                print ("Step: ", step)
                print ("tf.argmax(softmax_hy,1):", sess.run(tf.argmax(softmax_hy,1), feed_dict={X: test_data_X, Y: test_data_Y}))
                correct_prediction = tf.equal(tf.argmax(softmax_hy,1), test_data_labels)
                accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
                accuracy = accuracy.eval({X: test_data_X, Y: test_data_Y})
                if (accuracy > highest_test_accuracy):
                    highest_test_accuracy = accuracy
                print ("Accuracy: \t", accuracy)
                print ("test_data_labels: \t", test_data_labels)

        correct_prediction = tf.equal(tf.argmax(softmax_hy,1), test_data_labels)
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
        accuracy = accuracy.eval({X: test_data_X, Y: test_data_Y})
        if (accuracy > highest_test_accuracy):
            highest_test_accuracy = accuracy
        print ("Accuracy: ", accuracy)
        
        print ("End of one epoch: Accuracy: ", accuracy)
        test_accuracies.append(highest_test_accuracy)
        print ("test_accuracies: ", test_accuracies)
        
    print ("")
```

```python
mean_test_accuracy = np.mean(test_accuracies)
print ("mean_test_accuracy: ", mean_test_accuracy)
std_test_accuracy = np.std(test_accuracies)
print ("std_test_accuracy: ", std_test_accuracy)
```
