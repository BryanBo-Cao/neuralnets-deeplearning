```python
def get_data():
    image_data = np.zeros((50,32,32,3))
    for i in range(50):
        filename = "assign4_part3_data/" + str(i) + ".jpg"
        image = img.imread(filename)
        image_data[i] = image
        imgplot = plt.imshow(image)
        plt.show()
    return image_data
```
